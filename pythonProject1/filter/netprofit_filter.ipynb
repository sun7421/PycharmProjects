```python

import pandas as pd
a = pd.read_csv('value_filter.csv')
```

```python
x=a.iloc[0:,1:]
x
```

```python
netprofit=0
```

```python
number=x.shape[0]
number 
```

```python
 y=x.iloc[0,16]
y
```

```python
i=0
e=0
while e< number:
    y=x.iloc[i,16]
    if y > netprofit :
         i+=1
         e+=1
    else : 
         x = x.drop(e,axis=0)
         x.reset_index(drop=True) # Reset index
         e+=1   
print("finish")
    
```

```python
x.to_csv('netprofit_scan.csv')
x
```
