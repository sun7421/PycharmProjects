```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
x = pd.read_excel('stocksun.xlsm')
```

```python
number=x.shape[0]
number 
```

```python
freefote = 35
```

```python
 y=x.iloc[0,18]
y
```

```python
i=0
e=0
while e< number:
    y=x.iloc[i,18]
    if y<=freefote :
         i+=1
         e+=1
    else : 
         x = x.drop(e,axis=0)
         x.reset_index(drop=True) # Reset index
         e+=1   
print("finish")
    
```
