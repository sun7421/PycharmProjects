```python
stock_ema=x.iloc[0:,25] #first stock in ema
stock_ema
```

```python
#เทสEMA
```

```python
 number=x.shape[0]
 y=x.iloc[0,0]
y
```

```python
i=0
e=0
while e <number:
  y=x.iloc[i,0]
  end = datetime.today()
  duration= timedelta(days=365)
  start=end-duration
  s = web.DataReader(y+".BK", 'yahoo', start, end)
  close_array=s.iloc[0:,3:-2]
  close_array['EMA'] = talib.EMA(close_array['Close'], timeperiod =set_ema0)
  close_data = close_array.iloc[-1,0]
  ema_data0 =close_array.iloc[-1,1]
  stock_ema0 = ema_data0
  close_array['EMA'] = talib.EMA(close_array['Close'], timeperiod =set_ema1)
  ema_data1 =close_array.iloc[-1,1]
  stock_ema1 = ema_data1
  close_array['EMA'] = talib.EMA(close_array['Close'], timeperiod =set_ema2)
  ema_data2 =close_array.iloc[-1,1]
  stock_ema2 = ema_data2
  if close_data > stock_ema0 and close_data > stock_ema1 and close_data > stock_ema2 :
   x.iloc[i,25]=stock_ema0
   x.iloc[i,26]=stock_ema1
   x.iloc[i,27]=stock_ema2
   i+=1
   e+=1
  else  : 
   x = x.drop(e,axis=0)
   x.reset_index(drop=True) # Reset index
   e+=1
print("finish")       
```

```python
x.to_csv('finish_scan.csv')
x
```
